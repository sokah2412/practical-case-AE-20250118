```python
cee = pd.read_csv("data/cee_ops.csv")
journeys = pd.read_csv("data/journeys.csv", usecols=['driver_uid', 'status', 'route_id', 'cancel_reason', '_id'])
routes = pd.read_csv("data/routes.csv", usecols=['driver_uid', 'status', '_id'])
```

```python
validated_journeys = journeys[journeys.status == 'VALIDATED']
routes_with_covoit = routes.merge(validated_journeys, left_on='_id', right_on='route_id', how='left')
routes_with_covoit.head()
```

```python
nb_covoit_by_routes_id = routes_with_covoit.groupby('_id_x')\
    .count()[['_id_y']]\
    .reset_index()\
    .rename(columns={'_id_x': '_id', '_id_y': 'nb_covoits'})
nb_covoit_by_routes_id
```

```python
nb_covoit_by_routes = routes.merge(nb_covoit_by_routes_id, left_on="_id", right_on="_id")
nb_covoit_by_routes.sort_values('nb_covoits')
```

```python
nb_covoit_by_driver = nb_covoit_by_routes[nb_covoit_by_routes.nb_covoits > 0].groupby('driver_uid').count()[['_id']].reset_index()
nb_covoits_frequency = nb_covoit_by_driver.groupby('_id').count().reset_index()\
    .rename(columns={"_id": "nb_covoits_by_driver", "driver_uid": "nb_driver"})
nb_covoits_frequency['cumsum'] = nb_covoits_frequency.loc[::-1, 'nb_driver'].cumsum()[::-1]
nb_covoits_frequency.plot(x='nb_covoits_by_driver', y="cumsum")
```
